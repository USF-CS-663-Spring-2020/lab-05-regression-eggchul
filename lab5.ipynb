# Read toluca.csv file
#### target: lot size
#### the amount of hours of work denpends on the lot size
#### therefore, we analysis the linear regression relation between lot size and work hours

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

toluca_df = pd.read_csv("toluca.csv")
credit_df = pd.read_csv("credit.csv").iloc[:,1:7]
```

```python
toluca_df 
```

```python
toluca_df.plot(kind="scatter", x="workHours", y="lotSize")
plt.show()
```

```python
plot = sns.lmplot(y = "lotSize", x = "workHours", data = toluca_df)
plt.show()
```

```python
r1 = sns.residplot('workHours', 'lotSize', data = toluca_df)
r1.set(xlabel='Residuals', ylabel= 'Lot Size')
#randomly distribute, varience decreases
#help check up the linear regression plot we have
#residual plot should not be able to predict anyting
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(toluca_df.drop('lotSize', axis = 1), toluca_df.iloc[:,0])
lin_reg
#what it comes out is not a dataFrame. it is a data Series. one clomun in the two dimensional graph
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
# we found the intercept of the linear regression
```

We can conclude that there is a linear regression relationship between work hours and lot size. 'workHours' is suitable for SLR(Simple Linear Regression).

#### "Limit"|"Income""Rating""Age""Cards""Education"

```python
credit_df
```

## Credit Limit vs Income

```python
credit_df.plot(kind="scatter", x="Income", y="Limit")
plt.show()
```

```python
plot_income_limit = sns.lmplot(y = "Limit", x = "Income", data = credit_df)
```

```python
r_income_limit = sns.residplot('Income', 'Limit', data = credit_df)
r_income_limit.set(xlabel='Residuals of Income', ylabel= 'Limit')
plt.show()
```

```python
lin_reg_income_limit = LinearRegression()
lin_reg_income_limit.fit(credit_df.drop(['Limit','Rating','Age','Cards','Education'], axis = 1), credit_df.iloc[:,1])
lin_reg_income_limit
```

```python
lin_reg_income_limit.coef_
```

```python
lin_reg_income_limit.intercept_
```

From the LR graph above, we could find that there is a connection between credit limit and income.
However, when we check on the residual grapgh, the distribution of those spots does not spread evenly.
Therefore, 'income is not suitable for SLR'
