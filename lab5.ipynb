# Read toluca.csv file
#### target: lot size
#### the amount of hours of work denpends on the lot size
#### therefore, we analysis the linear regression relation between lot size and work hours

```python
import pandas as pd
toluca_df = pd.read_csv("./toluca.csv")
credit_df = pd.read_csv("credit.csv").iloc[:,1:7]
```

```python
toluca_df 
```

```python
toluca_df.plot(kind="scatter", x="lotSize", y="workHours")
```

```python
import seaborn as sns
plot = sns.lmplot(y = "workHours", x = "lotSize", data = toluca_df)
```

```python
r1 = sns.residplot('lotSize', 'workHours', data = toluca_df)
r1.set(xlabel='Lot Size', ylabel= 'Residuals')
#randomly distribute, varience decreases
#help check up the linear regression plot we have
#residual plot should not be able to predict anyting
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(toluca_df.drop('workHours', axis = 1), toluca_df.iloc[:,1])
lin_reg
#what it comes out is not a dataFrame. it is a data Series. one clomun in the two dimensional thing
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
# we found the intercept of the linear regression??
```
